```python
def flag_df(df):
  if (df.proportion <= 0.6):
    return 'Research'
  elif (0.6 < df.proportion < 1):
    return 'Undergrad'
  else:
    return 'Undergrad Only'

dffalse = df1[df1.is_highlvl == False]

dffalse['Flag'] = dffalse.apply(flag_df, axis = 1)
dffalse
```

```python
# Divirsity within post bachelor education vs bachelor education

#currdf[currdf.unitid == 175139]
df2 = dffalse.groupby('unitid').mean().reset_index()
df2['Flag'] = df2.apply(flag_df, axis = 1)
df_bc = df2[['unitid', 'proportion', 'Flag']]
df_bc
```

```python
dffinal = dfa.merge(df_bc, on=['unitid'], how='right')
dffinal.groupby(['unitid', 'year', 'cipcode', 'majornum']).sum()
```
